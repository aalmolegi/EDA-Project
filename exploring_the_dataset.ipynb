# Hypothses

## 1. The price of a house does not depend significantly on whether it is located near other expensive houses.

### Correlations between ```price``` and ```sqft_living15```
### Correlations between ```price``` and ```sqft_lot15```

````python
cor_Price_living15 = df['price'].corr(df['sqft_living15'])
cor_Price_lot15 = df['price'].corr(df['sqft_lot15'])

print(f"Correlation between price and sqft_living15: {cor_Price_living15:.2f}")
print(f"Correlation between price and sqft_lot15: {cor_Price_lot15:.2f}")
````

### Scatter plot for ```price``` and ```sqft_living15```
### Scatter plot for ```price``` and ```sqft_lot15```

````python
# Scatter plot for price vs. sqft_living15
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sqft_living15', y='price', data=df)
plt.title('Price vs. Square Footage of Neighboring Houses')
plt.xlabel('Square Footage of Neighboring Houses')
plt.ylabel('Price')
plt.show()

# Scatter plot for price vs. sqft_lot15
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sqft_lot15', y='price', data=df)
plt.title('Price vs. Lot Size of Neighboring Houses')
plt.xlabel('Lot Size of Neighboring Houses')
plt.ylabel('Price')
plt.show()
````

### Line plot for ```price``` and ```sqft_living15```
### Line plot for ```price``` and ```sqft_lot15```

````python
# Bin the 'sqft_living15' column into intervals
bins_living15 = np.arange(df['sqft_living15'].min(), df['sqft_living15'].max() + 400, 400)  # Create bins of width 500
df['sqft_living15_binned'] = pd.cut(df['sqft_living15'], bins=bins_living15, include_lowest=True)

# Calculate the average price for each bin
average_prices_by_bin = df.groupby('sqft_living15_binned')['price'].mean().reset_index()

# Create line plot
plt.figure(figsize=(10, 6))
plt.plot(average_prices_by_bin['sqft_living15_binned'].astype(str), average_prices_by_bin['price'], 
         marker='o', linestyle='-', color='b')

# Customize the plot
plt.title('Average Price vs. Square Footage of Neighboring Houses')
plt.xlabel('Square Footage of Neighboring Houses')
plt.ylabel('Average Price')
plt.xticks(rotation=90)
plt.grid(True)

# Show the plot
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
````

````python
# Bin the 'sqft_lot15' column into intervals
bins_lot15 = np.arange(df['sqft_lot15'].min(), df['sqft_lot15'].max()+50000, 50000)  # Create bins of width 1000
df['sqft_lot15_binned'] = pd.cut(df['sqft_lot15'], bins=bins_lot15, include_lowest=True)

# Calculate the average price for each bin
average_prices_by_bin = df.groupby('sqft_lot15_binned')['price'].mean().reset_index()

# Create line plot
plt.figure(figsize=(10, 6))
plt.plot(average_prices_by_bin['sqft_lot15_binned'].astype(str), average_prices_by_bin['price'], 
         marker='o', linestyle='-', color='g')

# Step 4: Customize the plot
plt.title('Average Price vs. Lot Size of Neighboring Houses')
plt.xlabel('Lot Size of Neighboring Houses')
plt.ylabel('Average Price')
plt.xticks(rotation=90) 
plt.grid(True)

# Show the plot
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
````

````python
# Extract month from the date column
df['month'] = pd.to_datetime(df['date']).dt.month

# Group by month and calculate average price
monthly_prices = df.groupby('month')['price'].mean()

# Plot the trend
plt.figure(figsize=(8, 6))
monthly_prices.plot(marker='o')
plt.title('Average Price by Month')
plt.xlabel('Month')
plt.ylabel('Average Price')
plt.grid(True)
plt.show()
````

````python
# Group by yr_built and calculate average price
avg_price_by_year = df.groupby('yr_built')['price'].mean().reset_index()
avg_price_by_year = avg_price_by_year.sort_values(by='price').head(20)

# Plot average price by yr_built
plt.figure(figsize=(12, 6))
plt.plot(avg_price_by_year['yr_built'], avg_price_by_year['price'], marker='o', linestyle='-', color='b')

# Highlight the cheapest year
cheapest_year = avg_price_by_year.iloc[0]['yr_built']
cheapest_price = avg_price_by_year.iloc[0]['price']
plt.scatter(cheapest_year, cheapest_price, color='red', s=100, label=f'Cheapest Year: {cheapest_year}')
plt.title('Average House Price by Effective Year Built')
plt.xlabel('Effective Year Built')
plt.ylabel('Average Price')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
````
